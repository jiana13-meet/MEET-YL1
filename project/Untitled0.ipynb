I got this data from http://lib.stat.cmu.edu/datasets/IQ_Brain_Size

```python
import csv

f = open('brainiq.csv', 'rb')
r = csv.reader(f)
IQ = []
Brain_Size = []

for row in r:
    row=row[0].replace("/t"," ")
    row=row.split()
    row=row[7:]
    row=map(float,row)
    IQ.append(row[0]IQ)
    Brain_Size.append(row[1])
f.close()
from pylab import *
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('The relationship between the brain size and the IQ')
ax.set_xlabel('Brain Size')
ax.set_ylabel('IQ')
ax.scatter(IQ, Brain_Size)
```

The x axes repreresents Brain Size,
The y axes represents IQ,
There is no relationship between these two elements, because there is no slope.
